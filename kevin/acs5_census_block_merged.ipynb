```python
def wrangle_acs5(src):
    df1 = pd.read_json(src)
    df1.columns = df1.iloc[0]
    df1 = df1[1:].copy()
    df1['block_group'] = df1.state.map(str) + df1.county.map(str) + df1.tract.map(str) + df1['block group'].map(str)
    df1=df1.drop(columns=['state','county','tract','block group']).set_index('block_group').astype('float')
    return df1
```
